# K nearest neighbor photo-z

For a very simple photo-z estimator, let's choose some input features, e.g. all magnitudes, all colors, etc..., and construct a KD-tree of our training data.  This enables us to quickly find the "closest" points (sklearn will return these, assuming a simple Euclidean distance metric in the input data space).  We can then assign each test galaxy a redshift based on its nearest neighbors, so a fairly direct mapping of the color space to redshift by assuming that the training data accurately summarizes the color -> redshift relationship.

````python
from sklearn.neighbors import KDTree
````

Here is where we define which data we will use as input to our PZ algorithm.  Columns included in `kdtree_cols` will be output to a numpy 2D array that will constitute the input to the sklearn kd-tree code:

````python
bands = ["u", "g", "r", "i", "z", "y"]
kdtree_cols = []
for band in bands:
    kdtree_cols.append(f"mag_{band}_lsst")
print(kdtree_cols)
````

````python
trainmagdata = traindf[kdtree_cols].to_numpy()
testmagdata = testdf[kdtree_cols].to_numpy()
````

we also need to separately pull out the true redshifts as a separate array, as we will need these values to predict our redshifts:

````python
sztrain = traindf['redshift'].values
sztest = testdf['redshift'].values
````

Now let's construct our tree, which we do with sklearn's `KDTree` class:

````python
magkdtree = KDTree(trainmagdata, leaf_size=10)
````

For a set of input data consisting of the same parameters that we used to set up our tree,`KDTree`'s `query` method will return the distances and indices for the `k` "closest" (using a Euclidean distance metric in the input params) training-set objects.  As a simple photo-z estimate we can take the **mean  value** of the redshifts for these k-nearest-neighbors of the training-set as the redshift estimate for each galaxy in the input data.  We can define a simple function to do just that, feeding in the KDTree, the testdata which we wish to predict redshifts for, the array of true redshifts for the training data, and a value `k`, which the user can set:

````python
def k_nearest_neighbor(kdtree, testdata, trainsz, k):
    dists, idxs = kdtree.query(testdata, k=k)
    matchpz = trainsz[idxs]
    meanz = np.mean(matchpz, axis=1)
    return meanz
````

Let's just do a straight single nearest-neighbor run, i.e. by setting k=1 we will simply assign every galaxy in the test-set the redshift of the nearest galaxy in the training set, then plot up the results:

````python
k1_pzs = k_nearest_neighbor(magkdtree, testmagdata, sztrain, 1)
````

````python
plt.figure(figsize=(8,8))
plt.scatter(sztest, k1_pzs, s=2, c='k')
plt.plot([0,3],[0,3],'r--');
plt.xlabel("redshift")
plt.ylabel("k=1 photo-z");
````

Looks very good!  Now, let's try increasing the number of neighbors:

````python
k_val = 5
meanzs = k_nearest_neighbor(magkdtree, testmagdata, sztrain, k_val)
````

````python
plt.figure(figsize=(8,8))
plt.scatter(sztest, meanzs, s=2,c='k')
plt.plot([0,3],[0,3], 'r--');
plt.xlabel("true redshift")
plt.ylabel(f"K={k_val} nearest neighbor")
````

# Simple Neural Network
try using scikit-learn's neural network to compute a photo-z.  We'll set which columns to feed into the net with the list `nnet_cols`:

If you want to try using colors + 1 magnitude rather than six magnitudes, you can do so in the cell below by switching which line is commented out.  You can also set any combination from the input data that you want, e.g. include magnitude errors as well.

````python
#nnet_cols = ["mag_u_lsst", "mag_g_lsst", "mag_r_lsst", "mag_i_lsst", "mag_z_lsst", "mag_y_lsst"]
nnet_cols = ["ug", "gr", "ri", "iz", "zy", "mag_i_lsst"]
````

## Activity: try other combinations of columns, e.g. the colors, colors + 1 magnitude, colors + all magnitudes, etc...

````python
# pull out the relevant data
nnet_train = traindf[nnet_cols].to_numpy()
sztrain = traindf['redshift'].values
nnet_test = testdf[nnet_cols].to_numpy()
sztest = testdf['redshift'].values
````

sklearn's neural net regressor has a couple of parameters, look at the documentation for `MLPRegressor` for details, we will set som fairly standard values and not go into detail on what they mean.  The MLPRegressor docs page is here: <br>
https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html

````python
import sklearn.neural_network as sknn
simplenn = sknn.MLPRegressor(hidden_layer_sizes=(12, 12),
                                     activation='tanh', solver='lbfgs',
                                     max_iter=2000)
````

# Question: what might you do in order to choose "good" values for these algorithm free parameters?

````python
#let's train our neural net:
simplenn.fit(nnet_train, sztrain)
````

````python
#we can now predict a redshift with our trained model for our test data:
nnet_pz = simplenn.predict(nnet_test)
````

````python
plt.figure(figsize=(8,8))
plt.scatter(sztest, nnet_pz, s=1, c='k')
plt.plot([0,3],[0,3], 'r--')
plt.xlabel("true redshift")
plt.ylabel("neural net estimate")
````
